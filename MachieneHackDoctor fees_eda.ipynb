```python

import pandas as pd
import numpy as np
import os
os.chdir("D:\Data Science\Machinehack\Final Participant Data Folder")
###Data Ingestion for the train and test files
train = pd.read_excel("D:/Data Science/Machinehack/Final Participant Data Folder/Final_Train.xlsx")
print(train.head())
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

sns.set_style("whitegrid")
# sns.set(rc={'axes.facecolor':'white', 'figure.facecolor':'white'})
#plt.style.use('ggplot')
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
sns.distplot(train.Fees, bins=50)
plt.title('Original')

plt.subplot(1,2,2)
sns.distplot(np.log1p(train.Fees), bins=50)
plt.title('Log transformed')

plt.tight_layout()
```
