Load model based on inception arch and pretrained weights . 

```python
#used by model as loss function
def triplet_loss(y_true, y_pred, alpha = 0.2): 
    anchor, positive, negative = y_pred[0], y_pred[1], y_pred[2]
    pos_dist = K.sum(K.square(anchor-positive), axis=-1)
    neg_dist = K.sum(K.square(anchor-negative),axis=-1)
    loss = pos_dist-neg_dist+alpha
    loss = K.sum(K.maximum(loss,0))
    return loss

#load model
def load_model():
    model = faceRecoModel(input_shape=(3, 96, 96))
    model.load_weights(MODEL_WEIGHTS)
    model.compile(optimizer = 'adam', loss = triplet_loss, metrics = ['accuracy'])    
    return model

FaceRecoModel = load_model()
FaceRecoModel.summary()
```

```python
#Common routines
#load image and correct color order
def load_image(path):
    img = cv2.imread(path, 1)
    # OpenCV loads images with color channels in BGR order. changing to RGB
    return img[...,::-1]
    
def get_face_cv2(image):
    global faceCascade
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    facebb = [1,1,1,1]
    faces = faceCascade.detectMultiScale(
            gray,
            scaleFactor=1.1,
            minNeighbors=5,
            minSize=(30, 30),
            flags=cv2.CASCADE_SCALE_IMAGE
        )
    if faces is not None and len(faces)>0:
        #for face in faces:
        (x, y, w, h) = faces[0]
        #print("image data",image.shape,x,y,x+w,y+h)    
        faceimg = image[x:x+w,y:y+h,:]
        faceimg = cv2.resize(faceimg, (96, 96))
        faceimg = np.transpose(faceimg, (2,0,1)) #model takes channel first so transpose faceimg from channel last
        facebb  = [x,y,w,h]           
         
    else:
        faceimg=None
    
    return faceimg,facebb 

#use dlib library to extract and align face
def get_face(image):
    bb = alignment.getLargestFaceBoundingBox(image)
    if bb is None : #skip if no face is detected
        return None
    faceimg = alignment.align(96, image, bb, landmarkIndices=AlignDlib.OUTER_EYES_AND_NOSE)
    faceimg = np.transpose(faceimg, (2,0,1)) #model takes channel first so transpose as source image will be channel last
    facebb=[1,1,1,1]
    w = bb.right()-bb.left()
    h = bb.bottom()-bb.top()
    facebb= [[bb.left(),bb.top(),w,h]]    
    return faceimg,facebb

#Detect face  and encode to Vector using model
def encode_image(image):
    faceimg,facebb =get_face(image)    
    if faceimg is not None :        
        faceimg = (faceimg / 255.).astype(np.float32)        
        enc = FaceRecoModel.predict(np.expand_dims(faceimg, axis=0))[0]
    else:
        enc=None
    return enc,facebb 

#square distance between 2 encoded vectors
def distance(vec1, vec2):
    return np.sum(np.square(vec1 - vec2))

#one subfolder per person. Images will be used for identity check
def load_dict_initial(dirname):
    X=[]
    Y=[]
    i=0
    for sub_dir in os.listdir(dirname):        
        print(sub_dir)
        for idfile in glob.glob(dirname+sub_dir+"/*.jpg"):
            #print(idfile)
            image=load_image(idfile)
            enc,_ = encode_image(image)
            if enc is not None:
                X.append(enc)
                Y.append(sub_dir)
    X= np.vstack(X)
    Y= np.asarray(Y)
    return X,Y

#save all id data as pickle file. expensive to load it from image each time
def load_dict():
    with open(ID_PICKLE, 'rb') as f:
        [a,b] = pickle.load(f)
        print("loaded identity",a.shape,b.shape)
    return a,b

#save the generated id vectors for future use
def save_dict(x,y):
    with open(ID_PICKLE, 'wb') as f:
        pickle.dump([x,y], f)
    f.close()

#load encoded data and corresponding labels
X,Y=load_dict()
#save_dict(X,Y)
```

Identify optimal threshold distance between the clusters. This is done to maximize accuracy 

```python
#find accuracy score to identify distance threshold
def identify_threshold(X,Y):
    dist = [] # Distance between pairs
    compare = [] # 1 if same identity, 0 otherwise
    num = len(X)
    for i in range(num - 1):
        for j in range(1, num):
            dist.append(distance(X[i], X[j]))
            compare.append(1 if Y[i] == Y[j] else 0)        
    dist = np.array(dist)
    compare = np.array(compare)
    thresholds = np.arange(0.01, 2, 0.01)
        
    acc_scores = [accuracy_score(compare, dist < t) for t in thresholds]
    idx = np.argmax(acc_scores) # check max accuracy score at which threshold index        
    threshold = thresholds[idx] # use this for filter criteria
    accuracy = acc_scores[idx]  # accuracy_score(identical, distances < optimal_threshold)
    print("Accuracy : ",accuracy," at threshold",threshold)

    return threshold

threshold=identify_threshold(X,Y)
```

Two options:
<li>Basic Identification : Use the encoded informationa and compare new image encodings and filter by threshold</li>
<li>Classifier : If there are multiple source images for each person in database, train a classifier. To avoid classifer false positve for unknown ids, do a round of filter to check for threshold match</li>

```python
def create_classfier(X,Y):
    svc = LinearSVC()
    #to start with use one third of the input images for training classifier; rest for testing the accuracy
    train_idx = np.arange(len(Y)) % 3 == 0
    test_idx = np.arange(len(Y)) % 3 != 0
    Xtrain = X[train_idx]
    ytrain = Y[train_idx]
    Xtest = X[test_idx]    
    ytest = Y[test_idx]
    id_classifier=svc.fit(Xtrain,ytrain)
    classifier_accuracy = accuracy_score(ytest, id_classifier.predict(Xtest))
    print("classifier accuracy",classifier_accuracy)
    return id_classifier

#check list for min dist and return if mindist < threshold
def identify_basic(image):
    enc,bb = encode_image(image)
    mindist=100
    idx=0
    for i in range(X.shape[0]):  
        dist = distance(enc,X[i])
        if  dist < mindist :            
            mindist=dist
            idx=i
    if mindist < threshold :
        return Y[idx],bb
    else: 
        return -1,None
    
#identify using classifier and check if the idetified id is meeting threshold check
def identify(image):
    enc,box = encode_image(image)
    enc = enc.reshape(1,enc.shape[0])
    y = id_classifier.predict(enc) #classifier might predict but will not handle -ve cases. so check with threshold
    l = np.where(Y == y)
    x = X[l] #check that the distance for this group if id is within the threshold. if not, will return unknown
    count = 0
    enc = enc.reshape((128))
    dist=100
    for idx in range(x.shape[0]):  
        dist = distance(enc,x[idx])
        if  dist <= threshold :            
            count +=1
            break
    if count > 0 :
        print("correct prediction by classifier",y,"distance check ",dist,  " threshold is ",threshold)
        return np.asscalar(y),box
    else: 
        print("incorrect prediction by classifier",y, dist,"no actual match within threshold",threshold)
        return -1,box
    
def identify_file(fname):
    img=load_image(fname)
    return identify(img)

id_classifier = create_classfier(X,Y)
identify_file("D:/Python/Scripts/FaceRecog/Identityv1/17/image_0280.jpg")
```

```python
#start training
class LossLayer(Layer):
    def __init__(self, alpha, **kwargs):
        self.alpha = alpha
        super(LossLayer, self).__init__(**kwargs)
    
    def triplet_loss(self, inputs):
        anchor, positive, negative = inputs
        posdist = K.sum(K.square(anchor-positive), axis=-1)
        negdist = K.sum(K.square(anchor-negative), axis=-1)
        return K.sum(K.maximum(posdist - negdist + self.alpha, 0), axis=0)
    
    def call(self, inputs):
        loss = self.triplet_loss(inputs)
        self.add_loss(loss)
        return loss

#build the model
input_shape = (3,96, 96)
uinp = Input(input_shape)
pinp = Input(input_shape)
ninp = Input(input_shape)
uemb = FaceRecoModel(uinp)
pemb = FaceRecoModel(pinp)
nemb = FaceRecoModel(ninp)
loss_layer = LossLayer(alpha=0.2, name='triplet_loss')([uemb, pemb, nemb])
trainmodel = Model([uinp, pinp, ninp], loss_layer)
trainmodel.summary()
trainmodel.compile(loss=None, optimizer='adam')
data=get_train_batch()
a_batch = data[:,0,:,:,:]
p_batch = data[:,1,:,:,:]
n_batch = data[:,2,:,:,:]
trainmodel.fit([a_batch , p_batch, n_batch])
```
