```python
tran_X = train
import re
def extract_city(z):
    check =re.search(', (.+)', z)
    if check:
        return(check.group(1))
    else:
        return 'NA'
hi ="Kakkanad, Ernakulam"
#print(extract_city(hi))
tran_X['City'] = tran_X.Place.apply(lambda x: extract_city(str(x)))
```

```python
##converting the ratings to number
import re
def convert_num(text):
    check =re.search('(.+?) years experience', text)
    if check:
        return(check.group(1))
    else:
        return 0
print(convert_num("1 years experience" ))
tran_X['Exp_years'] = tran_X.Experience.apply(lambda x: convert_num(x))
tran_X['Exp_years'] = tran_X['Exp_years'].astype(str).astype(int)
```

```python
print(tran_X.head())
```

```python
## convert city to Grade 1 and 2
def City_grade(city):
    if city in ('Ernakulam','Coimbatore','Thiruvanthapuram','NA'):
        return "Grade2"
    else:
        return "Grade1"
print(City_grade("Emakulam"))
tran_X['City_Grade'] = tran_X.City.apply(lambda x: City_grade(str(x)))
```

```python
tran_X.head()
```

```python
##converting the ratings to number
def feedback(z):
    
    check =re.search('% (.[0-9]+?) ', z)
    if check:
        return(check.group(1))
    else:
        return 0
hi="97% 60 Feedback Defence Colony, Delhi"
##print(feedback(hi))
tran_X['Feedback'] = tran_X.Miscellaneous_Info.apply(lambda x: feedback(str(x)))
    

```

```python
print(tran_X.head())
```

```python
##convert Rating to number
def convert_perc(text):
    check =re.search('(.+?)%', text)
    if check:
        return(check.group(1))
    else:
        return 0
tran_X['Rating_num'] = tran_X.Rating.apply(lambda x: convert_perc(str(x)))
tran_X['Rating_num'] = tran_X['Rating_num'].astype(str).astype(int)
print(tran_X.head())
```
